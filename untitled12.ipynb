```python
np.random.seed(1)
```

```python
# 1000 INTERGERS BETWEEN 0 AND 50
x = np.random.randint(0, 50, 1000)
```

```python
print(x)
```

```python
#Positive Correlation with some noise
y = np.random.normal(0, 10, 1000) - x 
```

```python
print(y)
```

```python
np.corrcoef(x, y)
```

```python
import pandas as pd

df = pd.DataFrame({'a':np.random.randint(0, 50, 1000)})
df['b'] = df['a'] + np.random.normal(0, 10, 1000) #positive correlation
df['c'] = 100 - df['a'] + np.random.normal(0, 5, 1000) #negative correlation
df['d'] = np.random.randint(0, 50, 1000) #not correlated with 'a'

df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.xticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
